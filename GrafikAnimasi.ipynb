```python
from tkinter import *
from matplotlib.figure import Figure
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
import numpy as np


root=Tk()
root.title('Belajar menggambar grafik')
root.resizable(width=False,height=False)
Lebar=1000
Tinggi=650
canvas=Canvas(root,width=Lebar,height=Tinggi,bg='lightblue')
canvas.pack()
def bikingrafik():
    a=float(aInput.get())
    b=float(bInput.get())
    c=float(cInput.get())
    xmin=int(xminInput.get())
    xmak=int(xmakInput.get())
    
    x=np.linspace(np.pi*xmin,np.pi*xmak,1001)
    
    if yInput.get()==yvalues[0]:
        y=a*np.sin(b*x)+c
        ax.plot(x,y)
        
    if yInput.get()==yvalues[1]:
        y=a*np.cos(b*x)+c
        ax.plot(x,y)
        
    if yInput.get()==yvalues[2]:
        y=np.sin(b*x)*np.cos(c*ax)
        ax.plot(x,y)
        
    if yInput.get()==yvalues[3]:
        x1=(x-c)
        x2=x1**2
        x3=b*x2
        y=a*np.exp(x3)
        ax.plot(x,y)
        
    canvasGrafik.draw()
    
def resetgrafik():
    global ax
    
    ax.clear()
    ax.set_title('Grafik Fungsi')
    ax.set_xlabel('x dalam pi')
    ax.set_ylabel('y')
    ax.grid(True)
    
    canvasGrafik.draw()

frameGrafik=Frame(root,bg='white')
frameGrafik.place(relx=1,rely=0.5,relwidth=0.7,relheight=1,anchor='e')

frameInput=Frame(root,bg='#074447')
frameInput.place(relx=.275,rely=0.5,relwidth=0.25,relheight=.7,anchor='e')
yvalues=('y = a sin(bx) + c','y = a cos(bx) + c','y = a sin(bx)cos(cx)','y = a exp(b(x-c)^2)')

pilihlabel=Label(frameInput,bg='#074447',text='Pilih fungsi yang akan di gambar',fg='white')
pilihlabel.place(relx=.4,rely=0,relwidth=1,relheight=.1,anchor='n')

yInput=Spinbox(frameInput,values=yvalues)
yInput.place(relx=0.4,rely=.1,relwidth=0.5,relheight=.05,anchor='n')

berilabel=Label(frameInput,bg='#074447',text='Masukkan nilai a, b dan c',fg='white')
berilabel.place(relx=.4,rely=.2,relwidth=1,relheight=.1,anchor='n')

alabel=Label(frameInput,bg='#074447',text='a=',fg='white')
alabel.place(relx=.1,rely=.3,relwidth=0.1,relheight=.1,anchor='n')
aInput=Entry(frameInput)
aInput.place(relx=0.2,rely=.33,relwidth=0.1,relheight=.05,anchor='n')
aInput.insert(0,1)

blabel=Label(frameInput,bg='#074447',text='b=',fg='white')
blabel.place(relx=.32,rely=.3,relwidth=0.1,relheight=.1,anchor='n')
bInput=Entry(frameInput)
bInput.place(relx=0.42,rely=.33,relwidth=0.1,relheight=.05,anchor='n')
bInput.insert(0,1)

clabel=Label(frameInput,bg='#074447',text='c=',fg='white')
clabel.place(relx=.54,rely=.3,relwidth=0.1,relheight=.1,anchor='n')
cInput=Entry(frameInput)
cInput.place(relx=0.64,rely=.33,relwidth=0.1,relheight=.05,anchor='n')
cInput.insert(0,0)

xlabel=Label(frameInput,bg='#074447',text='x dari',fg='white')
xlabel.place(relx=.1,rely=.5,relwidth=0.25,relheight=.1,anchor='n')
xlabel=Label(frameInput,bg='#074447',text='ke',fg='white')
xlabel.place(relx=.35,rely=.5,relwidth=0.25,relheight=.1,anchor='n')
xminInput=Entry(frameInput)
xminInput.place(relx=0.2,rely=.55,relwidth=0.1,relheight=.05,anchor='w')
xminInput.insert(0,-2)
xmakInput=Entry(frameInput)
xmakInput.place(relx=0.4,rely=.55,relwidth=0.1,relheight=.05,anchor='w')
xmakInput.insert(0,2)
xlabel=Label(frameInput,bg='#074447',text='dalam satuan pi',fg='white')
xlabel.place(relx=.7,rely=.5,relwidth=0.35,relheight=.1,anchor='n')

enter_button=Button(frameInput,text='Enter',command=bikingrafik)
enter_button.place(relx=0.4,rely=.65,relwidth=0.2,relheight=.06,anchor='w')

reset_button=Button(frameInput,text='Reset',command=resetgrafik)
reset_button.place(relx=.4,rely=.8,relwidth=0.2,relheight=.06,anchor='w')

f=Figure()
ax=f.add_subplot(111)


ax.set_title('Grafik Fungsi')
ax.set_xlabel('x dalam pi')
ax.set_ylabel('y')
ax.grid(False)

canvasGrafik = FigureCanvasTkAgg(f,frameGrafik)
canvasGrafik.get_tk_widget().place(relwidth=1,relheight=1)
canvasGrafik.draw()


root.mainloop()
```
